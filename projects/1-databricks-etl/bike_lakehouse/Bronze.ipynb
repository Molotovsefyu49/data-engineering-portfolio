# Define Ingestion Configuration

```python
# Base path for raw files
BASE_PATH = "/Volumes/workspace/bronze/source_systems"

# Files to ingest into bronze layer
FILES_TO_LOAD = [
    {
        "file_path": "source_crm/cust_info.csv",
        "table_name": "workspace.bronze.crm_cust_info"
    },
    {
        "file_path": "source_crm/prd_info.csv",
        "table_name": "workspace.bronze.crm_prd_info"
    },
    {
        "file_path": "source_crm/sales_details.csv",
        "table_name": "workspace.bronze.crm_sales_details"
    },
    {
        "file_path": "source_erp/CUST_AZ12.csv",
        "table_name": "workspace.bronze.erp_cust_az12"
    },
    {
        "file_path": "source_erp/LOC_A101.csv",
        "table_name": "workspace.bronze.erp_loc_a101"
    },
    {
        "file_path": "source_erp/PX_CAT_G1V2.csv",
        "table_name": "workspace.bronze.erp_px_cat_g1v2"
    }
]
```

#Ingestion into Bronze tables

```python


def load_csv_to_table(file_path, table_name):

    df = (
        spark.read
        .format("csv")
        .option("header", True)
        .option("inferSchema", True)
        .load(file_path)
    )

    (
        df.write
        .mode("overwrite")
        .saveAsTable(table_name)
    )

    print(f"Loaded {file_path} -> {table_name}")


for file in FILES_TO_LOAD:

    full_path = f"{BASE_PATH}/{file['file_path']}"
    table_name = file["table_name"]

    load_csv_to_table(full_path, table_name)
```
